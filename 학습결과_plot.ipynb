# VALID 4만장

```python
f = open("valid_0605_1823.log", "r", encoding="utf8")

y = []
for line in f.readlines():
    if line.find("xent: ") != -1:

        word_list = line.split('xent: ')
        y.append(word_list[1][:4])
            

f.close()
x = [i for i in range(1000, 60001, 1000)]
y = y[:-1]

print(y)
print(min(y))
print(x)
```

```python
for i in range(len(y)):
    y[i] = float(y[i])
    
plt.plot(x, y)
plt.xlabel("step")
plt.ylabel("xent")
plt.ylabel
plt.legend("Extractive model")
plt.title("Extractive Model Training xent")

print(min(y))
```

# Train 6만장 

```python
f = open("train_0604_2056.log", "r", encoding="utf8")

y = []
for line in f.readlines():
    if line.find("xent: ") != -1:
        word_list = line.split('/')
        if word_list[0][-3:] == "000":
            y.append(line.split("xent: ")[1][0:4])
            
f.close()
x = [i for i in range(1000, 60001, 1000)]


print(min(y))
print(len(x))
print(len(y))

```

```python
for i in range(len(y)):
    y[i] = float(y[i])
    
plt.plot(x, y)
plt.xlabel("step")
plt.ylabel("xent")
plt.ylabel
plt.legend("Extractive model")
plt.title("Extractive Model Training xent")

print(min(y))
```
